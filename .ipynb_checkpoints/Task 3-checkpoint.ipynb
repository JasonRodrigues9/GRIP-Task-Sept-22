## Heat-map for co-relation

```python
data.corr()
plt.figure(figsize=(10,7))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

## Profit vs Sales

```python
plt.figure(figsize=(10,8))
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

## Sub-category wise profit

```python
plt.figure(figsize=(12,9))
category_wise_profit=data.groupby(['Sub-Category'])['Profit'].sum().nlargest(60)
category_wise_profit.plot.barh()
```

## Profit v/s Sales sub-category wise 

```python
top_subcategory_sales = data.groupby("Sub-Category").Sales.sum().nlargest(n=100)
top_subcategory_profit = data.groupby("Sub-Category").Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
top_subcategory_sales.plot(kind='bar',figsize=(12,9),fontsize=12)
top_subcategory_profit.plot(kind='bar',figsize=(12,9),fontsize=12,color='red')
plt.ylabel('Total Sales & Profit')
plt.xlabel('Sub-Category')
plt.title('Sub-Category Sales v/s Profit')
plt.show()
```

## Bifurcation based on regional sales

```python
plt.figure(figsize=(6,4))
plt.title("Region")
plt.pie(data['Region'].value_counts(),labels=data['Region'].value_counts().index,autopct="%2.2f%%")
plt.show()
```

## State wise bifurcation

```python
plt.figure(figsize=(15,15))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```
